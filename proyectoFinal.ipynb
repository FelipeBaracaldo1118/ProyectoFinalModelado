#¿Qué es la Interpolación Spline Cúbica?
La interpolación spline cúbica es una técnica de interpolación que conecta puntos de datos con una serie de polinomios de tercer grado (cúbicos), uno para cada intervalo entre puntos consecutivos. #¿Por Qué el Spline Cúbico es la Mejor Opción?
Para este problema, el spline cúbico es preferible a otros métodos de interpolación debido a varias razones:

Suavidad y Realismo Físico:

La trayectoria de un balón sigue una curva suave y continua debido a las fuerzas físicas (gravedad, resistencia al aire). El spline cúbico asegura continuidad en la pendiente y en la curvatura (derivadas primera y segunda), lo cual modela mejor el comportamiento real del balón.
Evita el Problema de la Oscilación:

En interpolaciones polinómicas de alto grado (que se podrían aplicar si tuviéramos que usar un solo polinomio para todos los puntos), las curvas tienden a oscilar, especialmente en los extremos (el llamado "efecto de Runge"). Este comportamiento no es adecuado para trayectorias físicas. El spline cúbico evita este problema al utilizar polinomios de bajo grado en cada segmento.
Precisión en Cada Tramo:

El uso de un polinomio por cada intervalo permite capturar mejor la forma exacta de la trayectoria en cada tramo, proporcionando mayor precisión en la interpolación entre puntos de datos.
Cálculo de Puntos Específicos de Interés:

Como el spline cúbico genera una representación polinómica continua en cada tramo, es posible calcular fácilmente valores intermedios y derivadas, lo cual facilita encontrar la altura máxima (en el eje

#Conclusión
El spline cúbico proporciona una interpolación ideal para representar la trayectoria de un balón en movimiento, capturando la suavidad de la curva y permitiendo un análisis detallado de posiciones y derivadas en cada eje. Esto hace que sea el método preferido para este tipo de análisis espacial en 3D, donde los datos experimentales se ajustan a una curva continua que simula el comportamiento físico real de un objeto en movimiento.

```python
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize

# Paso 1: Definir los datos de entrada
# Datos de tiempo
t = np.array([0.00, 0.15, 0.30, 0.45, 0.60, 0.75, 0.90, 1.05, 1.20])
# Coordenadas en el espacio
x_t = np.array([0.00, 0.50, 1.00, 1.50, 1.80, 2.00, 1.90, 1.10, 0.30])
y_t = np.array([0.00, 4.44, 8.88, 13.31, 17.75, 22.19, 26.63, 31.06, 35.50])
z_t = np.array([0.00, 0.81, 1.40, 1.77, 1.91, 1.84, 1.55, 1.03, 0.30])

# Paso 2: Definir la función para calcular los coeficientes del spline cúbico
def cubic_spline(x, y):
    n = len(x) - 1  # Número de segmentos
    h = np.diff(x)  # Distancias entre los puntos de x
    alpha = np.zeros(n)  # Coeficientes de las diferencias

    # Calcular los coeficientes alpha
    for i in range(1, n):
        alpha[i] = (3/h[i]) * (y[i+1] - y[i]) - (3/h[i-1]) * (y[i] - y[i-1])

    # Inicializar matrices l, mu y z
    l = np.ones(n + 1)
    mu = np.zeros(n)
    z = np.zeros(n + 1)

    # Calcular l, mu y z
    for i in range(1, n):
        l[i] = 2 * (x[i+1] - x[i-1]) - h[i-1] * mu[i-1]
        mu[i] = h[i] / l[i]
        z[i] = (alpha[i] - h[i-1] * z[i-1]) / l[i]

    # Condiciones de frontera
    l[n] = 1
    z[n] = 0
    c = np.zeros(n + 1)  # Coeficientes c
    b = np.zeros(n)      # Coeficientes b
    d = np.zeros(n)      # Coeficientes d

    # Calcular b y d
    for j in range(n - 1, -1, -1):
        c[j] = z[j] - mu[j] * c[j + 1]
        b[j] = (y[j + 1] - y[j]) / h[j] - h[j] * (c[j + 1] + 2 * c[j]) / 3
        d[j] = (c[j + 1] - c[j]) / (3 * h[j])

    return b, c, d  # Devolver coeficientes b, c y d

# Paso 3: Definir la función para evaluar el spline cúbico
def evaluate_spline(x, y, b, c, d, x_eval):
    n = len(x) - 1  # Número de segmentos
    spline_values = np.zeros_like(x_eval)  # Valores del spline

    # Evaluar el spline en los puntos x_eval
    for i in range(n):
        mask = (x_eval >= x[i]) & (x_eval <= x[i + 1])  # Condición para el intervalo
        if np.any(mask):
            dx = x_eval[mask] - x[i]  # Diferencia entre x_eval y el punto x[i]
            spline_values[mask] = y[i] + b[i] * dx + c[i] * dx**2 + d[i] * dx**3

    return spline_values  # Devolver los valores evaluados

# Paso 4: Calcular los coeficientes para las coordenadas z(t)
b_z, c_z, d_z = cubic_spline(t, z_t)

# Paso 5: Evaluar el spline cúbico en un rango de t
t_eval = np.linspace(0, 1.20, 100)  # Crear un rango de tiempo
z_eval = evaluate_spline(t, z_t, b_z, c_z, d_z, t_eval)  # Evaluar el spline en z(t)

# Evaluar el spline para x(t) y y(t) también
b_x, c_x, d_x = cubic_spline(t, x_t)
b_y, c_y, d_y = cubic_spline(t, y_t)

x_eval = evaluate_spline(t, x_t, b_x, c_x, d_x, t_eval)  # Evaluar spline en x(t)
y_eval = evaluate_spline(t, y_t, b_y, c_y, d_y, t_eval)  # Evaluar spline en y(t)

# Paso 6: Encontrar el tiempo donde la altura es máxima
def find_max_height_time(b, c, d, t):
    def negative_spline(t_max):
        return -evaluate_spline(t, z_t, b, c, d, np.array([t_max]))  # Maximizar la altura

    # Minimizar la función negativa para encontrar el máximo
    result = minimize(negative_spline, x0=0.75, bounds=[(0, 1.20)])
    return result.x[0], -result.fun  # Devolver tiempo y altura máxima

# Calcular el tiempo y la altura máxima
max_height_time, max_height = find_max_height_time(b_z, c_z, d_z, t)
print(f"Tiempo de máxima altura: {max_height_time:.2f} s, Altura máxima: {max_height:.2f} m")

# Paso 7: Determinar la altura z del balón al cruzar la barrera
# La barrera se ubica a y = 9 m
t_barrera = np.interp(9, y_t, t)  # Interpolación lineal para encontrar el tiempo
z_barrera = evaluate_spline(t, z_t, b_z, c_z, d_z, np.array([t_barrera]))[0]  # Evaluar el spline
print(f"Altura del balón al cruzar la barrera: {z_barrera:.2f} m")

# Paso 8: Determinar la desviación máxima (dx/dt=0)
# Encontrar el punto más alto de la trayectoria
x_max = x_t[np.argmax(z_t)]  # Posición x en la altura máxima
y_max = max_height  # Altura máxima
z_max = evaluate_spline(t, z_t, b_z, c_z, d_z, np.array([max_height_time]))[0]  # Altura z en la altura máxima
print(f"Desviación máxima en la posición: (x, y, z) = ({x_max:.2f}, {y_max:.2f}, {z_max:.2f})")

# Paso 9: Imprimir los polinomios de cada segmento
print("\nPolinomios del Spline Cúbico para z(t):")
for i in range(len(b_z)):
    print(f"Segmento {i}: z(t) = {z_t[i]} + {b_z[i]}(t - {t[i]}) + {c_z[i]}(t - {t[i]})^2 + {d_z[i]}(t - {t[i]})^3")

# Paso 10: Graficar la trayectoria del balón en 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_t, y_t, z_t, 'ro', label='Datos originales')  # Graficar los datos originales
ax.plot(x_eval, y_eval, z_eval, label='Spline Cúbico', color='b')  # Graficar el spline cúbico
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Z (m)')
ax.set_title('Trayectoria del Balón')
ax.legend()
plt.show()

```
